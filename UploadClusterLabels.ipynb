```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function
from __future__ import unicode_literals

import pandas as pd
import json
import requests
import sklearn.cluster as skc
from sklearn.preprocessing import normalize
```

```python
ADDRESS = "http://192.168.1.11"
PORT = 8000

get_queries_url = "{}:{}/iris/retrieve-participant-queries".format(ADDRESS, PORT)
upload_cluster_labels_url = "{}:{}/iris/insert-cluster-ids".format(ADDRESS, PORT)
```

```python
participant_id = 0
mission_id = 1
```

```python
data = {
    'participant_id': participant_id,
    'mission_id': mission_id
}
req = requests.post(url=get_queries_url, data=data)
data = pd.DataFrame(req.json()['data'])
data
```

```python
if len(data) <= 0:
    print("No queries to upload.")
```

```python
data = data[(data["type"] == "visual-detection") | (data["type"] == "audio-detection")]
```

```python
di = {
    "visual-detection":0,
    "audio-detection":1
}
```

```python
data.replace(di, inplace=True)
```

```python
mission_queries = data[["query_id", "confidence", "type"]].as_matrix()

model = skc.DBSCAN(eps=0.011, min_samples=4).fit(
    normalize(mission_queries[:, 1:]))

labels = model.labels_

print(labels)
```

```python
data.shape
```

```python
print(labels.shape)
```

```python
label_dict = dict(zip(data["query_id"], labels))
```

```python
data = {
    "cluster_assignments":json.dumps({str(k):str(v) for k, v in label_dict.items()[:100]})
}
```

```python
data["cluster_assignments"]
```

```python
req = requests.post(url=upload_cluster_labels_url, data=data)
```
