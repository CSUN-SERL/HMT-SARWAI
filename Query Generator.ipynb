```python
import pandas as pd
import random
import time 
```

```python
cols = ['q_start', 'q_service_time', 'q_type', 'q_conf', 'action']
```

```python
def q_generator():
    while True:
        q_type = random.randint(0, 6)
        q_conf = random.random()
        q_start = time.time()

        if q_conf < .33:
            action = 0
            q_service_time = random.randrange(6, 9)
        elif q_conf <.66 and q_conf >= .33:
            action = 1
            q_service_time = random.randrange(3, 6)
        else:
            action = 2
            q_service_time = random.randrange(1, 3)
        
        yield q_start, q_service_time, q_type, q_conf, action
```

```python
q_gen = q_generator()
```

```python
data = pd.DataFrame()
for col in cols:
    data[col] = ''
```

```python
cols = ['q_start', 'q_service_time', 'q_type', 'q_conf', 'action']
```

```python
num_queries = 200
```

```python
for x in range(num_queries):
    start, q_service_time, q_type, q_conf, action = next(q_gen)
    data = data.append(dict(zip(cols, [start, q_service_time, q_type, q_conf, action])), ignore_index=True)
    time.sleep(random.randrange(0, 2))
```
